## 문자열 인덱싱 & 슬라이싱 문제
문자열 인덱싱 또는 슬라이싱을 사용하여 년 월 일 날씨를 출력하시오
text = '20230601sunny'
- year = '2023'
- month = '06'
- day = '01'
- weather = 'sunny'

```python
text = '20230601sunny'

year = text[:4]
month = text[4:6]
day = text[6:8]
weather = text[8:]

print(year)
print(month)
print(day)
print(weather)
```

## 전화번호 형식 바꾸기 문제

010-1234-5678의 형식으로 저장된 전화번호를 '-' 를 제거해 01012345678의 형식으로 바꿔보시오  
- 문자열 내장 함수를 사용하면 해결 가능
- 문자열 슬라이싱을 써도 상관 X

```python
phone_number = '010-1234-5678'
print(phone_number.replace('-', ''))
```

## 문자열 분할 문제 (Split)
카카오톡의 대화 정보는 다음과 같이 저장된다  
"2023년 1월 2일 오후 9:52, 유연휘 : 일단"  
위 문장을, 날짜, 시간, 이름, 내용별로 분할해보자.

```python
# 대화 정보
kakao = "2023년 1월 2일 오후 9:52, 유연휘 : 일단"

# 쉼표와 콜론을 기준으로 자르고
date_time, name_content = kakao.split(", ")

# 자른걸 다른 기준으로 다시 자르기
date, time = date_time.split(" 오후 ")
name, content = name_content.split(" : ")

print(f"날짜: {date}")
print(f"시간: 오후 {time}")
print(f"이름: {name}")
print(f"내용: {content}")
```
