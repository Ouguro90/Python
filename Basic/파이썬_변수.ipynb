# 파이썬의 변수

데이터가 저장되어 있는 메모리 공간

- 파이썬에서는 할당연산자 ``=`` 로 변수를 정의할 수 있다
  - 변수명 = 변수값
  - 보통 수학공식 처럼 `변수명은 변수값` 이라고 생각할 수 있지만
    프로그래밍에서의 ``=`` 는 다르다
  - 같다가 아닌 변수의 메모리에 저장된 값을 가르킨다
  - 변수값은 변경될 수 있다
```
a = 10
b = 20
```


변수를 할당하게 되면 변수를 통해 저장된 데이터를 몇번이고 재사용할 수 있다
코드의 효율성이 높아지는 것이다

## 변수 생성시 주의사항

### 1. 가능한 영문으로 설정
- 한글을 사용해도 실행에는 크게 문제가 없지만 한글 변수를 실행하기 위해
  별도의 인코딩이 필요한 경우가 종종 있기때문에 가능한 영문을 사용

### 2. 첫번째 문자는 소문자로 설정 권장
- 첫번째 문자를 대문자로 해도 크게 상관은 없지만   
  대문자는 파이썬에서 Class를 표현할 때 주로 사용되기 때문에 혼동을 줄 수 있다
- 중간 대문자는 상관 없음

### 3. 예약어 및 함수명 사용 금지
- if, for 등의 예약어나 존재하는 함수명을 변수명으로 사용하면,  
  오류가 발생하거나 해당 함수명을 사용하지 못하게 되는 경우가 발생하기 때문에  
  사용하지 않도록 한다

### 4. 특수문자 및 공백문자 사용 금지
- 언더바 ``_`` 를 제외한 특수문자와 공백문자는 오류를 발생시키기 때문에  
  변수명에 포함되지 않도록 주의
- 맨뒤의 공백은 상관없음
- 맨앞의 공백문자는 권장하지 않음 **(Error 뜨는 경우 있음)**

````python
# 맨앞에 특수문자 사용 시 문법 오류 발생
$a = 10
print($a)
````

### 5. 맨앞에 숫자 사용 금지
- 숫자를 중간이나 뒤에 붙이는건 상관이 없으나 맨앞은 오류가 발생한다

````python
# 맨앞에 숫자 사용 시 문법 오류 발생
1a = 10
print(1a)
````
