# 제어문

프로그램의 흐름을 제어하는 구문.  

**조건문**, **반복문**, **예외처리** 구문이 있다

## 조건문 (if, elif, else)

프로그래밍에서 조건을 판단하여 해당 조건에 맞는 상황을 수행하는 데 쓰는 것

### **if**문

조건이 참일 때 특정 블록의 코드를 실행하는 기본적인 조건문

if문의 기본 구조

```python
if 조건:        # 조건은 참(True) 혹은 거짓(False)를 반환하는 표현식 이어야 한다
  실행할 코드   # 조건이 참일 경우 실행되는 코드
```

````python
x = 10

if x > 5:
  print('x는 5보다 큽니다')
````

### **elif**문

여러 조건을 연속적으로 체크할 때 사용되는 조건문,  
``if`` 조건이 거짓일 경우,  다음 ``elif`` 조건을 검사하는 방식  

여러개의 ``elif`` 사용 가능

elif문의 기본 구조

```python
if 조건1:
  실행할 코드1

# if의 조건이 거짓일 때만 검사
# if의 조건이 참이면 나머지 elif, else는 실행 X

elif 조건2:
  실행할 코드2  

elif 조건3:
  실행할 코드3

else:
  실행할 코드
```

### **else**문

``if``나 ``elif`` 조건들이 모두 거짓일 때 실행되는 블록  

else는 선택 사항이며, ``if``나 ``elif`` 뒤에 위치해야 한다

else 사용법

```python
if 조건:
  실행될 코드

else:           # if 조건이 거짓이면 실행
  실행될 코드
```

## 반복문 (while, for)

### while문

주어진 조건이 참(True)인 동안 코드를 반복 실행



while문의 기본 구조
```python
while 조건:
  조건이 참인 경우 실행되는 문장
```

````python
# while문 외부에서 조건을 위한 변수 선언
num = 1

# while 조건:
while num <= 5:
  # 조건이 참인 경우 실행되는 문장
  print(num)
  # 조건을 변경시키는 문장을 항상 while문 내부에 포함시켜야함
  # (무한 루프 방지)
  num += 1
````

#### 연습문제  

n보다 작거나 같은 3의 배수 합 구하기 예제  

요구사항  
- 사용자로부터 정수 n을 입력받는다
- while 반복문을 사용하여 1부터 n까지의 숫자 중 3의 배수만 더한 합계를 출력한다

````python
num = int(input())
a = 1
result = 0

while a <= num:
  if a % 3 == 0:
    result = a + result
  a += 1
print(result)
````

1부터 입력한 수까지 홀수만 출력하되, 100보다 크면 종료하는 예제  

요구사항
- 사용자로부터 정수 n을 입력받는다
- while 반복문을 사용하여 1부터 n까지의 숫자 중 홀수만 출력
- 단, 숫자가 100보다 크면 더 이상 출력하지 않고 반복문 종료
- continue는 짝수를 건너뛸 때 사용하고, break는 100 초과 시 종료할 때 사용

````python
num = int(input())
a = 0

while a <= num:
  a += 1
  if a % 2 == 0:
    continue
  elif a > 100:
    break
  print(a, end=' ')
````

### for문

while문과 비슷하게 진행되지만, 조금 더 직관적  

continue, break 문도 역시 사용 가능하다

for문의 기본구조
```python
for i in 반복가능한 객체(리스트, 튜플, 등):
  수행할 문장1
  수행할 문장2
  ...
```

````python
list1 = ['one', 'two', 'three', 'four', 'five']

for i in list1:
  print(i)
````

for문으로 두가지 이상의 값 제어

````python
a = [(1, 1), (2, 2), (3, 3)]

for (b, c) in a:
  print(b+c)
````

````python
students = [('철수', 85), ('영희', 92), ('민수', 78)]

for (name, score) in students:
  print(f'{name}의 점수는 {score} 입니다')
````

#### 연습문제

구구단 출력 예제  

9 x 1 = 9  
9 x 2 = 18  
9 x 3 = 27  
...  
9 x 9 = 81  

````python
for i in range(1, 10):
  for j in range(2, 10):
    print(f'{j} X {i} = {i*j}', end='\t')
  print()
````

학생 점수에 따라 합격/불합격 판별 예제

요구사항
- 학생들의 점수 리스트 scores = [15, 50, 46, 86, 94] 가 주어진다  
- for 반복문을 활용해 홀수 번째 학생만 합격/불합격을 판단
  1. num이라는 번호를 1번부터 증가
  2. 번호가 짝수일 경우 continue를 사용하여 건너뛴다
  3. 번호가 5가 되면 break를 사용하여 건너뛴다
  4. 그 외에는 점수가 50 이상이면 'x번 학생 합격!', 그렇지 않으면 'x번 학생 불합격!' 을 출력

````python
students = [15, 50, 46, 86, 94]

num = 0

for i in students:
  num += 1
  if num % 2 == 0:
    continue

  if num == 5:
    break

  if i >= 50:
    print('%d번 학생 합격!' %num)
  else:
    print('%d번 학생 불합격!' %num)
````

## 반복문 제어 (break, continue, pass)

### break

break문을 만나면 반복문을 즉시 종료

````python
for i in range(5):
  if i == 3:
    break
  print(i)
````

### continue

continue문을 만나면 조건을 따지지 않고 다음 반복으로 넘어감

````python
for i in range(5):
    if i == 3:
        continue
    print(i)
````

### pass

아무것도 하지 않고 넘어감 (자리만 차지)

````python
for i in range(5):
    if i == 3:
        pass  # 그냥 넘어감
    print(i)
````

## 예제 심화

요구사항  
- 학생들의 점수 리스트 scores = [15,50,46,86,94,100,51,60]
- for 반복문과 조건문을 활용하여 다음 작업을 수행
  1. 각 학생에게 번호(1부터)를 부여
  2. 홀수 번호 학생만 검사하여, 점수가 60점 이상이면 'x번 학생 합격!', 그렇지 않으면 'x번 학생 불합격!' 을 출력
  3. 불합격한 홀수 번호 학생들의 점수를 누적하여 평균을 계산
  4. 반복이 끝난 후, 불합격자들의 평균 점수를 '불합격 받은 학생의 평균 점수는 : 00 입니다' 형식으로 출력

````python
scores = [15,50,46,86,94,100,51,60]
failed = []
num = 0

result = 0

for i in scores:
  num += 1

  if num % 2 == 0:
    continue

  if i >= 60:
    print('%d번 학생 합격!' %num)
  else:
    print('%d번 학생 불합격!' %num)
    failed.append(i)
    print(failed)
    result += i


result = int(result / len(failed))
print(f'불합격 받은 학생의 평균 점수는 {result} 입니다')
````
