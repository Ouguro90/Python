## 문제 1

두 숫자를 입력받아 더한 결과를 반환하는 함수를 작성한다.

- 함수 이름: `add_numbers`

## 문제 2

정수를 입력받아 짝수인지 홀수인지 판별하는 함수를 작성한다.

- 함수 이름: `check_even_odd`
- 매개변수: 정수 하나
- 반환값: 문자열 "짝수입니다." 또는 "홀수입니다."

### 예시

출력: 짝수입니다.

출력: 홀수입니다.

### 추가 : 짝수 / 홀수 구분하여 리스트에 저장


```python
even_list = []
odd_list = []

def check_even_odd(a):
  if a % 2 == 0:
    even_list.append(a)
    print("짝수입니다")
  else:
    odd_list.append(a)
    print('홀수입니다')

num = int(input())
check_even_odd(num)

print(even_list)
print(odd_list)
```

## 문제 3

주어진 세 숫자 중 가장 큰 수를 반환하는 함수를 작성한다.

- 함수 이름: `max_of_three`
- 매개변수: 세 개의 숫자
- 반환값: 가장 큰 숫자

### 예시
입력: 5, 8, 2  
출력: 8

### 추가 : 별도의 내장함수 사용하지 않고 구현

```python
def max_of_three(a, b, c):
    max_num = a
    if b > max_num:
        max_num = b
    if c > max_num:
        max_num = c
    return max_num


num1 = int(input())
num2 = int(input())
num3 = int(input())

max_of_three(num1, num2, num3)
```

## 문제 8

숫자 리스트를 입력받아, 0보다 큰 숫자만 골라서 새로운 리스트로 반환하는 함수를 작성한다.

- 함수 이름: `filter_positive_numbers`
- 매개변수: 숫자 리스트 하나
- 반환값: 양수만 담긴 리스트

### 예시
입력: [-2, 0, 5, 9, -1]  
출력: [5, 9]


```python
lst = []
new_lst = []

for _ in range(5):
  num = int(input())
  lst.append(num)

def filter_positive_numbers(num_lst):
  for i in num_lst:
    if i > 0:
      new_lst.append(i)
  return new_lst

filter_positive_numbers(lst)
```

## 문제 9

문자열을 입력받아, 단어별로 분리한 리스트를 반환하는 함수를 작성한다.

- 함수 이름: `split_into_words`
- 매개변수: 문자열 하나
- 반환값: 단어 리스트

### 예시
입력: "Python is fun"  
출력: ["Python", "is", "fun"]

### 추가 : 별도의 내장함수 사용하지 않고 구현

```python
words = []

def split_into_words(s):
  word = ""


  for char in s:
      if char != " ":        # 공백이 아니면 현재 단어에 추가
            word += char
      else:
          if word:           # 공백 만나기 전까지 모은 단어 저장
                words.append(word)
                word = ""
  if word:
        words.append(word)

  return words

str1 = input()

split_into_words(str1)
```

## 문제 11

정수 리스트를 입력받아, 리스트 안에서 중복된 숫자만 모아 리스트로 반환하는 함수를 작성한다.

- 함수 이름: `find_duplicates`
- 매개변수: 숫자 리스트 하나
- 반환값: 중복 숫자 리스트

### 예시
입력: [1, 2, 3, 2, 4, 1]  
출력: [1, 2]


```python
num_list = []

def find_duplicates(num_list):
  num_check = []
  num_duplicates = []


  for i in num_list:
    if i not in num_check:
      num_check.append(i)
    elif i not in num_duplicates:
      num_duplicates.append(i)

  num_duplicates.sort()
  return num_duplicates


for _ in range(6):
  num = int(input())
  num_list.append(num)

find_duplicates(num_list)
```

## 문제 12

두 개의 문자열을 입력받아, 두 문자열이 아나그램인지 판별하는 함수를 작성한다.

아나그램: 같은 알파벳을 사용해서 순서만 바꾼 단어
- 함수 이름: `is_anagram`
- 매개변수: 문자열 두 개
- 반환값: True 또는 False

### 예시
입력: ("listen", "silent")  
출력: True

입력: ("hello", "world")  
출력: False


```python
word1_list = []
word2_list = []

def is_anagram(word1, word2):
  for i in word1:
    word1_list.append(i)

  for j in word2:
    word2_list.append(j)

  word1_list.sort()
  word2_list.sort()

  if word1_list == word2_list:
    print('True')
  else:
    print('False')

word1 = input()
word2 = input()

is_anagram(word1, word2)
```

```python
def is_anagram(str1, str2):
    str1 = str1.lower()
    str2 = str2.lower()

    # 문자 길이 비교해서 다르면 애너그램 아니니까 False
    if len(str1) != len(str2):
        return False

    # 딕셔너리 선언
    freq = {}

    # 첫번째 단어 문자 빈도수 체크
    for char in str1:
        freq[char] = freq.get(char, 0) + 1

    # 두번째 단어 문자를 보며 첫번째 단어에서 기록한 빈도에서 하나씩 줄임
    for char in str2:
        # 문자가 없거나 개수가 0이 되면 애너그램이 아님
        if char not in freq or freq[char] == 0:
            return False
        freq[char] -= 1
    # 둘 다 통과 시 애너그램 맞음
    return True

word1 = input()
word2 = input()

is_anagram(word1, word2)
```
