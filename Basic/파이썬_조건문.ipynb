## 점수에 따른 등급 부여 문제

학생들에게 점수에 따라서 등급을 부여하려고 한다

90점 이상일 경우 A등급
80점 이상일 경우 B등급
70점 이상일 경우 C등급
70점 미만일 경우 D등급을 부여하는 코드 작성

```python
# 학생 점수 입력
score = int(input())

if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >= 70:
  print('C')
else:
  print('D')
```

## 더 큰 숫자 출력 문제

사용자에게 정수 값을 2개 입력받아, 더 큰 숫자를 출력하고,  
만약 같은 숫자라면 같습니다.  라는 문장을 출력

```python
# 사용자 정수 2개 입력

a = int(input())
b = int(input())

if a > b:
  print(a)
elif a < b:
  print(b)
else:
  print('같습니다')
```

## 숫자에 따른 단위 부여 문제

숫자가 1000 이상이면, 백단위를 생략하고 K를 붙인다  
숫자가 1000000 이상이면, 십만 단위를 생략하고 M을 붙인다  

1,000 -> K  
1,000,000 -> M  
1,000,000,000 -> G

- 입력 : 25000


```python
num = int(input())

if num > 1000000000:
  num = str(num)
  num = num.rstrip('0')
  print(num + 'G')

elif num > 1000000:
  num = str(num)
  num = num.rstrip('0')
  print(num + 'M')

elif num > 1000:
  num = str(num)
  num = num.rstrip('0')
  print(num + 'K')
```
